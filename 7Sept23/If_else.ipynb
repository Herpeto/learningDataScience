### Conditional Statement

Conditional statement merupakan statement untuk mengatur eksekusi program. Pengaturan tersebut berdasarkan boolean expression yang terdapat di setelah syntax **if** atau **elif**. Conditional statement di python harus diikuti oleh **else** sebagai kondisi default jika kondisi tidak terpenuhi

conditional.png

dalam penulisan conditional statement menggunakan python, perlu diperlukan indentasinya. karena python tidak seperti bahasa lain yang menggunakan tanda kurung untuk menandakan suatu blok kode. syntax if-else di bahasa python dapat dilihat digambar berikut:
<img src='img/Python-if-Statement-Syntax.png'>

| Kondisi | Output |
| --- | --- |
| Sales>1000 | Luar biasa! |
| Sales<1000 | Sedikit lagi! |

```python
### contoh conditional statement untuk kondisi tabel diatas
sales=500
if sales>1000:
    print('Luas biasa!')
else:
    print('sedikit lagi')
```

conditional statement juga bisa digunakan untuk mengatur jalannya program kondisi dengan menggunakan operator logical (and atau or). jumlah statement if yang diperlukan juga bisa menyesuaikan dengan jumlah percabangan yang ada.

| Sales | Profit | Output |
| --- | --- | --- |
| >1000 | >1000000 | Selamat, dapat bonus 100% |
| <1000 | >1000000 | Dapet bonus 75% |
| >1000 | <1000000 | Dapet bonus 50% |
| <1000 | <1000000 | Kurang beruntung|

```python
### contoh conditional statement dengan kondisi-kondisi diatas
sales=750
profit=1500000
if sales>1000 and profit>1000000:
    print("Selamat, dapet bonus 100%")
elif sales<1000 and profit>1000000:
    print("Dapet bonus 75%")
elif sales>1000 and profit<1000000:
    print("Dapet bonus 50%")
else:
    print("Kurang beruntung")
```

```python
###nested if
if sales>1000:
    if profit>1000000:
        print("Selamat, dapet bonus 100%")
    else:
        print("Dapet bonus 50%")
else:
    if profit>1000000:
        print("Dapet bonus 75%")
    else:
        print("Kurang beruntung")
```
