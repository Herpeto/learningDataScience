# Grouping dan Aggregating data

Setelah kita berhasil:
- load data (import data read_csv)
- edit data (drop kolom, drop baris, judul kolom, judul index)
- filter (ambil dengan kondisi)
- add data (nambah kolom)
- explore (mengambil data2 tertentu, describe, value_counts)

kalau kita mau analisa data, biasanya kita perlu teknik grouping

contoh: kita mau lihat dari tahun ke tahun berapa rata-rata pts yang di skor tim BOS per game

```python
# step 1: kita ambil data khusus tim BOS
df_bos = df_nba[df_nba["team_id"] == "BOS"]
df_bos
```

```python
# step 2: kita mau cari rata-rata pts yang di skor per game
df_bos["pts"].mean()
```

```python
# step 3: kita mau lihat rata2 pts per tahun, per year_id
df_bos.groupby("year_id")["pts"].mean()
```

kita tidak bisa lakukan groupby tanpa agregasi. 

```python
df_bos.groupby("year_id")
```

```python
df_bos.groupby("year_id").median()
```

```python
df_bos.groupby("year_id")["pts"].median()
```

groupby bisa lebih dari 1, misalnya kita mau lihat rata2 pts per tahun, untuk game2 yg menang dan kalah

```python
df_bos.groupby(["year_id", "game_result"])[["pts", "opp_pts"]].mean()
```

```python
# setiap tahunnya tim BOS menang berapa game, kalah berapa game
df_bos.groupby(["year_id", "game_result"])["game_id"].count()
```

agregasi termasuk:
- sum (total)
- mean (rata-rata)
- median (nilai tengah, 50% data ada di bawah dan diatas angka ini)
- min
- max

# Visualisasi

```python
df_nba.plot()
```

biasanya kita akan melakukan groupby, aggregation, dan perapihan data lainnya sebelum visualisasi

contoh: data rata-rata pts tim BOS dari tahun ke tahun bisa kit plot, line plot

```python
df_bos_pts_yby = df_bos.groupby("year_id")['pts'].mean()
df_bos_pts_yby
```

```python
df_bos_pts_yby.plot()
```

kita gunakan matplotlib untuk menambah judul di plot

```python
df_bos_pts_yby.plot(marker="o", figsize=(12,6))

plt.title("rata2 pts tim BOS per tahun")
```

ubah style graphic matplotlib nya

```python
# ganti style plot
plt.style.use("ggplot")

# bikin plot
df_bos_pts_yby.plot(marker="o", figsize=(12,6))

# nambah judul
plt.title("Rata2 skor tim BOS per tahun")

# ganti judul axis
plt.xlabel("Tahun")
plt.ylabel("Skor")

# save as png
plt.savefig("rata2 pts tim BOS per tahun.png")

plt.show()
```

Kita mau lihat juga untuk tim LAL dan NYK (3 tim yang paling banyak games nya) trend pts per tahun seperti diatas

```python
df_nba["team_id"].value_counts().head()
```

```python
df_bos = df_nba[df_nba["team_id"] == "BOS"]
df_lal = df_nba[df_nba["team_id"] == "LAL"]
df_nyk = df_nba[df_nba["team_id"] == "NYK"]

df_bos_pts_yby = df_bos.groupby("year_id")["pts"].mean()
df_lal_pts_yby = df_lal.groupby("year_id")["pts"].mean()
df_nyk_pts_yby = df_nyk.groupby("year_id")["pts"].mean()

df_3tim = pd.DataFrame({
    "BOS":df_bos_pts_yby, 
    "LAL":df_lal_pts_yby, 
    "NYK":df_nyk_pts_yby
    })

df_3tim
```

```python
# ganti style plot
plt.style.use("ggplot")

# bikin plot
df_3tim.plot(marker="o", figsize=(12,6))

# nambah judul
plt.title("Rata2 skor tim per tahun")

# ganti judul axis
plt.xlabel("Tahun")
plt.ylabel("Skor")

# save as png
plt.savefig("rata2 pts tim per tahun.png")

plt.show()
```
