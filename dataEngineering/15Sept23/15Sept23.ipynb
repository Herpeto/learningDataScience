```python
def value_to_int(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return int(x.replace('K', '')) * 1000
        return 1000
    if 'M' in x:
        if len(x) > 1:
            return int(x.replace('M', '')) * 1000000
        return 1000000
    return int(x)

test["Wage"] = test["Wage"].apply(value_to_int)
test["Wage"]
# type(test["Wage"].info())
```
