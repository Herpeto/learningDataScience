### Data Cleaning

```python
# Copying data into new Dataframe (df_filtered)
df_filtered = df.copy()
```

```python
# Manipulating Columns
# Deleting white space in front and end of the columns name

df_filtered.rename(columns=lambda x: x.strip(),inplace=True)
df_filtered.columns
```

```python
# Deleting Month Number, Month Name and Year
# It's because there is already Date Column
df_filtered.drop(['Month Number', 'Month Name','Year'],axis=1,inplace=True)
df_filtered.head()
```

```python
# To convert all the Money Data into Int
# Need to save the currency that we use in this dataframe

df_filtered['Sales'][0][0]
```

While trying to get the first char in Column for creating Table Currency, There is space infront and back of the data in the DataFrame

To fix that, we need to use .strip() on every column in dataframe

```python
# Clearing White Spaces in All Column in Dataframe
for cols in df_filtered.columns:
    if (df_filtered[cols].dtype == 'object'):
        df_filtered[cols] = df_filtered[cols].map(str.strip)
    else:
        pass

df_filtered.head()
```

```python
# Saving the Currency into it's own column
df_filtered['Use Currency'] = df_filtered['Sales'].astype(str).str[0]
df_filtered.head()
```

### Total Profit di Groupbykan pada Segment tanpa memasukkan data Discount

FROM financial_filtered

WHERE Discounts != 0

GROUP BY Segment;

### Statistik Sales per Negara
SELECT Country, ROUND(AVG(Sales),2) as 'Rata_Rata_Negara', ROUND(MIN(Sales),2) as 'Min_Negara', ROUND(MAX(Sales),2) as 'Max_Negara'

FROM financial_filtered

GROUP BY Country;
