### Python functions

Dalam suatu kode, terkadang syntax yang sama diperlukan dan digunakan secara berulang-ulang. Hal ini menyebabkan inefisiensi dan efektifitas kode jadi berkurang.

```python
### penggunaan conditional statement beberapa kali
angka=10
if angka%2==0:
    print("Genap")
else:
    print("Ganjil")

angka=15
if angka%2==0:
    print("Genap")
else:
    print("Ganjil")

angka=3
if angka%2==0:
    print("Genap")
else:
    print("Ganjil")
```

```python
# fungsi dengan dua parameter, untuk menentukan predikat kelulusan mahasiswa berdasarkan ipk dan lama studi
def get_status_kelulusan(ipk, lama_studi):
    if ipk>3.5 and ipk<=4.0 and lama_studi<=8:
        return "Cumlaude"
    elif ipk>3 and ipk<=4.0 and lama_studi<10:
        return "Sangat memuaskan"
    elif ipk>2.0 and ipk<=4.0 and lama_studi<=14 :
        return "Memuaskan"
    elif ipk>4.0 and lama_studi<=14:
        return "ipk tidak valid"
    elif ipk<=4.0 and lama_studi>=14:
        return "lama studi tidak valid"
    else:
        return "data dua-duanya tidak valid"

print(get_status_kelulusan(3.25,9))
print(get_status_kelulusan(3.7,7))
print(get_status_kelulusan(4.1,18))
```

### datetime library

library ini digunakan untuk memanipulasi tipe data date time dalam bahasa python.

#### now

now merupakan function yang akan mengembalikan datetime saat ini.

```python
### contoh now
import datetime
print(datetime.datetime.now())
```

#### strptime

merupakan syntax yang digunakan untuk mengubah string ke datetime sesuai dengan formast yang diinginkan. contohnya terlampir dalam gambar berikut strptime.PNG

format yang dapat digunakan adalah sebagai berikut timeSeriesFormat.png

```python
### contoh strptime
date_string='2020-06-13'
date_object=datetime.datetime.strptime(date_string, '%Y-%m-%d')
print(date_object)
```

```python
print(type(date_string))
print(type(date_object))
```

#### strftime

strftime merupakan kebalikan dari fungsi strptime, yaitu mengubah suatu datetime ke string sesuai dengan format yang diinginkan. proses ini sering dilakukan apabila kita ingin mendapatkan informasi spesifik dari suatu datetime

```python
### mendapatkan nama bulan dan hari
hari=datetime.datetime.strftime(date_object, '%B')
print("Bulan ini adalah: {0}".format(hari))
```

3. buat fungsi kalkulator scientific dengan parameter berupa perintah dan angkanya, dan mengembalikan nilai tersebut sebagai float. perlu diperhatikan bahwa string perintah bisa dimasukkan tanpa memperdulikan huruf besar dan kecilnya. string perintah dan operasi matematika yang dilakukan dapat dilihat di tabel bawah ini

| Perintah | Operasi |
| --- | --- |
| akardua | Mendapatkan akar pangkat dua dari bilangan |
| log10 | Mendapatkan logaritma basis 10 dari bilangan |
| pangkattiga | Mendapatkan pangkat tiga dari bilangan |
| log2 | Mendapatkan logaritma basis 2 dari bilangan |

```python
from math import *
def calculator(perintah, angka):
  if perintah.lower() == 'akardua':
    return (sqrt(angka))
  elif perintah.lower() == 'log10':
    return (log10(angka))
  elif perintah.lower() == 'pangkattiga':
    return (pow(angka,3))
  elif perintah.lower() == 'log2':
    return (log2(angka))
  else:
    return 0

print(calculator('Akardua',225))
print(calculator('log2',64))
print(calculator('pangkatTiga',7))
print(calculator('akarDua',729))
print(calculator('akar_Dua',100))
```
