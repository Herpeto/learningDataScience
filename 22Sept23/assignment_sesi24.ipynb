```python
department_jobs_sat = df.groupby(['Department','JobSatisfaction'])['JobSatisfaction'].count()
labels = []
for i in department_jobs_sat.index.get_level_values(0):
    if i in labels:
        pass
    else:
        labels.append(i)

print(labels)
print(department_jobs_sat['Research & Development'])
x = np.arange(len(labels))
width = 0.25
multiplier = 0

plt.bar(x, department_jobs_sat['Research & Development'], color = 'b',
        width = width, edgecolor = 'black', label='Human Resources')
plt.bar(x+width, department_jobs_sat['Human Resources'], color = 'g',
        width = width, edgecolor = 'black', label='Research & Development')
plt.bar(x+width+width, department_jobs_sat['Sales'], color = 'y',
        width = width, edgecolor = 'black', label='Sales')



# Add some text for labels, title and custom x-axis tick labels, etc.
plt.set_ylabel('Jumlah Orang')
plt.set_title('Satisfaction per Department')
plt.set_xticks(x + width/2, labels)
plt.legend(loc='upper left', ncols=3)
plt.set_ylim(0, 250)

plt.show()
```

```python
gender_jobs_sat = df.groupby(['Gender','JobSatisfaction'])['JobSatisfaction'].count()
gender_jobs_sat
```

```python
fig = plt.figure(figsize=(15,10), dpi=1600)

# First Column
ax1 = plt.subplot2grid((1,2),(0,0))
plt.pie(gender_jobs_sat['Female'],autopct='%1.2f%%')
plt.title('Job Satisfaction per Gender Female')
plt.legend(gender_jobs_sat['Female'].index)

# Second
ax1 = plt.subplot2grid((1,2),(0,1))
plt.pie(gender_jobs_sat['Male'],autopct='%1.2f%%')
plt.title('Job Satisfaction per Gender Male')
plt.legend(gender_jobs_sat['Male'].index)

plt.show()
```
