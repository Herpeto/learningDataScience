```python
# Nomor 3
# Hitung Jumlah Missing Values dan Persentasenya

# Jumlah Missing Values
df_building.isnull().sum()
```

```python
# Persentase Missing Values
df_building.isnull().mean()
```

```python
# Nomor 4
# Menghapus Kolom yang Persentase Missing Valuenya > 80%

df_building_filtered = df_building.copy()
above_8 = (df_building_filtered.isnull().mean() > 0.8)

df_building_filtered = df_building_filtered.drop((above_8[above_8] == True).index, axis=1)
df_building_filtered
```

```python
# Persentase Missing Values after Filtered
df_building_filtered.isnull().mean()
```

```python
# Jumlah yang diatas 0.8 setelah di filter
(df_building_filtered.isnull().mean() > 0.8).groupby((df_building_filtered.isnull().mean() > 0.8)).count()
```

```python
# Nomor 5
# Fungsi untuk mengisi kolom yang memiliki data kosong dengan mean dari kolom tersebut, median dari kolom tersebut, dan 'unknown'

def fillna_mean(df_kolom):
    mean = df_building_filtered[df_kolom].mean()
    df_building_filtered[df_kolom].fillna(mean, inplace=True)

def fillna_median(df_kolom):
    median = df_building_filtered[df_kolom].median()
    df_building_filtered[df_kolom].fillna(median, inplace=True)

def fillna_unknown(df_kolom):
    df_building_filtered[df_kolom].fillna('unknown', inplace=True)
```

```python
# Nomor 6
fillna_mean("Number of Existing Stories")

fillna_unknown("Street Suffix")

fillna_unknown("Proposed Use")
fillna_median("Proposed Units")

df_building_filtered.isnull().sum()
```

```python
# Nomor 7 
# Buat kolom baru Street, Street Name + Street Suffix

df_building_filtered["Street"] = df_building_filtered["Street Name"] + " "+ df_building_filtered["Street Suffix"]
df_building_filtered
```

```python
# Nomor 8
# Buat Kolom Year Month Day dari Kolom 'Permit Creation Date'

# Untuk mendapatkan Tahun, Bulan, dan Hari, kita harus merubah Kolom 'Permit Creation Date'
# Menuju type datetime
pd.to_datetime(df_building_filtered["Permit Creation Date"])

df_building_filtered["Year"] = pd.DatetimeIndex(df_building_filtered["Permit Creation Date"]).year
df_building_filtered["Month"] = pd.DatetimeIndex(df_building_filtered["Permit Creation Date"]).month
df_building_filtered["Day"] = pd.DatetimeIndex(df_building_filtered["Permit Creation Date"]).day

df_building_filtered
```

```python
# Nomor 9
# Ubah constr type 1, constr type 2, constr type 3, constr type 4 menjadi constr di 'Existing Construction Type Description'

df_building_filtered["Existing Construction Type Description"] = df_building_filtered["Existing Construction Type Description"].replace('constr type 1','constr')
df_building_filtered["Existing Construction Type Description"] = df_building_filtered["Existing Construction Type Description"].replace('constr type 2','constr')
df_building_filtered["Existing Construction Type Description"] = df_building_filtered["Existing Construction Type Description"].replace('constr type 3','constr')
df_building_filtered["Existing Construction Type Description"] = df_building_filtered["Existing Construction Type Description"].replace('constr type 4','constr')

df_building_filtered["Existing Construction Type Description"].value_counts()
```

```python
# Nomor 10
# Ubah angka pada values di kolom Description
df_building_filtered["Description"].value_counts()
```

```python
# Menghapus angka dari data yang ada di kolom Description
df_building_filtered["Description"].replace('\d+', '', regex=True, inplace=True)
df_building_filtered["Description"].value_counts()
```
