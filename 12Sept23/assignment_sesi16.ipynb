## Assignment Sesi 16 - Analyze Diabetes Dataset
This assignment will be using [this dataset (Diabetes Dataset)](https://raw.githubusercontent.com/stefkwan-h8/dataset/main/diabetes.csv) to be analyze, explore, manipulate, and filter.

This analysis will be using Python with Pandas and NumPY library

```python
# Nomor 1 - Import Library
import pandas as pd
import numpy as np
```

```python
# Nomor 2 - Download data
# Nomor 3 - Import data to Dataframe
df_diabetes = pd.read_csv("https://raw.githubusercontent.com/stefkwan-h8/dataset/main/diabetes.csv")
df_diabetes
```

```python
# Data yang dipilih untuk dihitung statistiknya adalah Kolom BMI
df_diabetes_bmi = df_diabetes["BMI"]
df_diabetes_bmi
```

```python
# Nomor 4 - Statistic of Some Column
arr_diabetes_bmi = np.array(df_diabetes_bmi) # Convert Dataframe BMI to Array

# Kalkulasi Statistik
diabetes_bmi_avg = arr_diabetes_bmi.mean() # AVG BMI
diabetes_bmi_median = np.median(arr_diabetes_bmi) # MEDIAN BMI
diabetes_bmi_std = arr_diabetes_bmi.std() # STD Deviat ion
diabetes_bmi_min = arr_diabetes_bmi.min() # Min of BMI
diabetes_bmi_max = arr_diabetes_bmi.max() # Max of BMI


print("Average of BMI from dataset : ",diabetes_bmi_avg)
print("Median of BMI from dataset : ",diabetes_bmi_median)
print("Standard Deviation BMI from dataset : ",diabetes_bmi_std)
print("Min value of BMI from dataset : ",diabetes_bmi_min)
print("Max value of BMI from dataset : ",diabetes_bmi_max)
```

```python
# Nomor 5 - Put new Column
df_diabetes_with_ratio = df_diabetes
df_diabetes_with_ratio.insert(len(df_diabetes.columns),'RatioBMItoBP',(df_diabetes["BMI"]/df_diabetes["BloodPressure"]))

df_diabetes_with_ratio
```

```python
# Nomor 6 - Input New Data (Randomly) to Dataframe
rng = np.random.default_rng()

new_row = [rng.integers(5), rng.integers(90,180), rng.integers(50,70), rng.integers(0,40), 0, rng.integers(20,40), np.random.rand(), rng.integers(20,50), 0, np.random.rand()]
df_diabetes_with_ratio.loc[len(df_diabetes_with_ratio)] = new_row
```

```python
df_diabetes_with_ratio
```

```python
# Nomor 7 - Filter Data tertentu
df_diabetes_1 = df_diabetes_with_ratio[df_diabetes_with_ratio["Outcome"] == 1]

# Nomor 8 - Simpan data filter menuju CSV baru
df_diabetes_1.to_csv("diabetes_outcome_1.csv")
```

```python
# Nomor 9 - Pattern tertentu
df_bmi_above_avg = df_diabetes_with_ratio[df_diabetes_with_ratio["BMI"] >= diabetes_bmi_avg]
df_bmi_above_avg_1 = df_bmi_above_avg[df_bmi_above_avg["Outcome"] == 1]
count_bmi_above_avg_1 = df_bmi_above_avg_1["Outcome"].count()
print("Jumlah orang yang memiliki BMI diatas rata-rata dan positif diabetes :",count_bmi_above_avg_1)

count_1 = df_diabetes_1["Outcome"].count()
print("Jumlah orang yang positif diabetes :",count_1)

ratio_bmi_above_avg = count_bmi_above_avg_1/count_1
print("\nRatio antara BMI diatas rata-rata dan positif diabetes dengan total orang positif : %.5f"%ratio_bmi_above_avg)
```
